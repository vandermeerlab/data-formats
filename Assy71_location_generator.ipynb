```python
#The locations values are assigned based on the images above
#The location keys are the AD channels from columsn 2 and 3 from probe_files/Assy77_P1_ChannelArrangement.pdf 

location_keys = [53,51,34,50,49,35,36,37,0,55,54,32,38,39,33,52,
                7,5,3,19,20,21,22,23,18,17,16,48,1,2,4,6,
                8,10,12,28,27,26,25,24,29,30,31,63,14,13,11,9,
                58,60,45,61,62,44,43,42,15,56,57,47,41,40,46,59]
location_values = [];
```

```python
seed1 = np.array([5.5,7.5])
seed2 = np.array([28,192.5])
y_offset = np.array([0,25])
x_offset = np.array([250,0])
for i in range(4):
    location_values.extend([seed1 + j * (y_offset) for j in range(8)])
    location_values.extend([seed2 - j * (y_offset) for j in range(8)])
    seed1 += x_offset
    seed2 += x_offset
```

```python
# Sanity check to see that the relative locations of the various channels match that of the images above
xcoords = [pos[0] for pos in location_values]
ycoords = [pos[1] for pos in location_values]
fig,ax = plt.subplots()
ax.axis('off')
for i in range(len(xcoords)):
    plt.text(xcoords[i]/200,ycoords[i]/200,location_keys[i])
```

```python
parent = {}
parent["channel_groups"] = {
        0: {
            'channels': (location_keys),
            'graph': [],
            'geometry': ([x.tolist() for x in location_values])
        }
}
```
